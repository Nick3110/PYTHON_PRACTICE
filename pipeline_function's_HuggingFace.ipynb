```python
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis')
classifier(['i have beeen waiting fro this course my whole life',
'i hate this so much'])
```

```python
classif=pipeline('sentiment-analysis')
classif(['i am happy',
            'i hate this'])
```

```python
classifier=pipeline('zero-shot-classification')
classifier(
    'As a poliyical leader I promise to build Mega Highway',
    candidate_labels=['education', 'politics', 'construction']
)
```

```python
unmasker = pipeline('fill-mask')
result = unmasker('hello what are you doing and what about your <mask>.', top_k=2)
print(result)
```

```python
ner = pipeline('ner',aggregation_strategy='first')
ner('my name is nikhil and i work at amazon in mumbai')
```
