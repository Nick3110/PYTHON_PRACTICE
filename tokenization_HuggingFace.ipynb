```python
from transformers import AutoTokenizer
```

```python
tokenzer = AutoTokenizer.from_pretrained('bert-base-uncased')
tokenzer.tokenize('lets try to tokenize! ')
```

```python
tokenzer = AutoTokenizer.from_pretrained('albert-base-v1')
tokenzer.tokenize('lets try to tokenize')
```

```python
tokenzer = AutoTokenizer.from_pretrained('bert-base-uncased')
tok=tokenzer.tokenize('lets try to tokenize!')
tok_id = tokenzer.convert_tokens_to_ids(tok)
```

```python
print(tok_id)
```

```python
tokenzer = AutoTokenizer.from_pretrained('bert-base-uncased')
tok = tokenzer.tokenize('lets try something new')
tok_id = tokenzer.convert_tokens_to_ids(tok)
tok_id
```

```python
#  special tokens adding
# prepare_model method
new_id = tokenzer.prepare_for_model(tok_id)
new_id
```
